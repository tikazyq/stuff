```python
from numpy import *
from pandas import *
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline

d = load_iris()
data = DataFrame(d.data, columns=d.feature_names).join(Series([d.target_names[i] for i in d.target], name='target'))

cat = list(set(data['target']))
scatter_matrix(data,cat_col='target',alpha=0.3,marker='o',diagonal='hist')
plt.show()
```
